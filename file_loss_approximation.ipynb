```python
import file_loss_analysis as fl
from pandas import DataFrame
```

```python
def launch_experiment(disks, chunk_count, threshold_recovery, spread_factor, experiment, delta, mean_reliability):
    """
    Args:
        disks (list):
        chunk_count (int):
        threshold_recovery (int):
        spread_factor (int):
        experiences (list):
        delta (double):
        mean_reliability (double):
        
    Returns:
    """
    
    # Approximated file loss probability
    reliability_mean_value = fl.probability_at_least_d_fail_equal_reliability(threshold_recovery, chunk_count, mean_reliability)
    print("----------------------------------------------------------------------")
    print("                  Approximated file loss probability                  ")
    print(reliability_mean_value)
    print("----------------------------------------------------------------------")
    
    # Exact reliability dataframe
    print("----------------------------------------------------------------------")
    print("                     Exact file loss probability                     ")
    reliabilities = DataFrame(fl.file_loss_delta_matrix(delta, mean_reliability, experiment, disks, chunk_count, spread_factor, 
                                                        threshold_recovery)
                              , columns=['Reliability'])
    
    # Histogram of reliabilities
    reliabilities.hist(bins=50)
    
    # Mean value and standard deviation of the dataframe reliabilities
    print "Mean value of reliabilities : ",reliabilities.mean()
    print("")
    print "Standard deviation of reliabilities : ",reliabilities.std()

    # relative error
    relative_error_list = abs(reliabilities - reliability_mean_value)/abs(reliabilities)
    
    # absolute error
    absolute_error_list = abs(reliabilities - reliability_mean_value)

    # relative error plot
    relative_error_list.plot(title='Relative errors plot')
    
    # relative error histogram
    relative_error_list.hist(bins=50)

    # absolute error plot
    absolute_error_list.plot(title='Absolute errors plot')
    
    # absolute error histogram
    absolute_error_list.hist(bins=50)
    
    print("----------------------------------------------------------------------")
```

```python
from numpy import mean

def deviation(delta, mean_reliability, experiment, disks, chunk_count, spread_factor, threshold_recovery):
    """
      Compute the deviation between the mean reliability and is approximate value 
      for a special delta.
      
    Args:
        delta (double):
        
    Returns:
        The standard deviation as a double.
    """
    
    # Approximated file loss probability
    reliability_mean_value = fl.probability_at_least_d_fail_equal_reliability(threshold_recovery, chunk_count, mean_reliability)
    

    # Exact probability
    reliabilities = [element for element in 
                     fl.file_loss_delta_matrix(delta, mean_reliability, experiment, disks, chunk_count, spread_factor, 
                                                        threshold_recovery)]
    
    return abs(mean(reliabilities) - reliability_mean_value)/abs(mean(reliabilities))
```

```python
from numpy import std

def std_vs_delta(delta, mean_reliability, experiment, disks, chunk_count, spread_factor, threshold_recovery):
    """
      Compute the standard for a special delta.
      
    Args:
        delta (double):
        
    Returns:
        The standard deviation as a double.
    """

    # Exact probability
    reliabilities = [element for element in 
                     fl.file_loss_delta_matrix(delta, mean_reliability, experiment, disks, chunk_count, spread_factor, 
                                                        threshold_recovery)]

    
    return std(reliabilities)
```

```python
delta_list = linspace(10**-6, 10**-3, 50)

m = 2000 # The number of disks in the system
disks = range(m)

n = 12 # number of chunks
d = 3 # number of chunks until the file is lost
spread_factor = 11 # spread factor

experience_count = 2000 # The number of experiments
experiment = range(experience_count)

mean_reliability = 0.999

std_list = [std_vs_delta(delta, mean_reliability, experiment, disks, n, spread_factor, d)
                      for delta in delta_list]

std_vs_delta_df = DataFrame(zip(delta_list, std_list), columns=['Delta', 'Standard Deviation'])
std_vs_delta_df.plot(x='Delta' ,y='Standard Deviation', style='o')
```
