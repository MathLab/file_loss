## Deviation between approximate value and mean value on experiment

```python
delta_list = linspace(10**-6, 10**-3, 50)

m = 2000 # The number of disks in the system
disks = range(m)

n = 12 # number of chunks
d = 3 # number of chunks until the file is lost
spread_factor = 12 # spread factor

experience_count = 10000 # The number of experiments
experiment = range(experience_count)

mean_reliability = 0.999

deviation_list = [fl.deviation(delta, mean_reliability, experiment, disks, n, spread_factor, d)
                      for delta in delta_list]

deviation_vs_delta = DataFrame(zip(delta_list, deviation_list), columns=['Delta', 'Deviation'])

deviation_vs_delta_log = DataFrame(zip([log(l) for l in delta_list], [log(l) for l in deviation_list]), columns=['Delta', 'Deviation'])
deviation_vs_delta_log.plot(x='Delta' ,y='Deviation', style='+')

slope, intercept, r_value, p_value, std_err = stats.linregress([log(l) for l in delta_list], [log(l) for l in deviation_list])
print(slope) # slope of the regression line
print("")
print(intercept) # intercept of the regression line
print("")
print(r_value) # correlation coefficient
print("")
print(p_value) # two-sided p-value for a hypothesis test whose null hypothesis is that the slope is zero
print("")
print(std_err) # Standard error of the estimate
```
